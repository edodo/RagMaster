```python
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
PDF_PATH = "/workspace/day_02/test.pdf"


def split_documents(docs: list[Document]) -> list[Document]:
    # LangChain Loader의 load_and_split()과 비슷한 역할
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=200,
    )

    return splitter.split_documents(docs)
```

```python
%%time

from pypdf import PdfReader


def load_pdf_with_pypdf(file_path: str) -> list[Document]:
    reader = PdfReader(file_path)

    docs = []

    for page_index, page in enumerate(reader.pages):
        text = page.extract_text() or ""

        docs.append(
            Document(
                page_content=text,
                metadata={
                    "source": file_path,
                    "page": page_index,
                    "parser": "pypdf",
                },
            )
        )

    return docs


pages = split_documents(load_pdf_with_pypdf(PDF_PATH))

print("청크의 수:", len(pages))

pages[10]
```

```python
%%time

import pymupdf


def load_pdf_with_pymupdf(file_path: str) -> list[Document]:
    pdf = pymupdf.open(file_path)

    docs = []

    for page_index in range(len(pdf)):
        page = pdf[page_index]
        text = page.get_text("text") or ""

        docs.append(
            Document(
                page_content=text,
                metadata={
                    "source": file_path,
                    "page": page_index,
                    "parser": "pymupdf",
                },
            )
        )

    pdf.close()

    return docs


pages = split_documents(load_pdf_with_pymupdf(PDF_PATH))

print("청크의 수:", len(pages))

pages[10]
```

```python
%%time

import pdfplumber


def load_pdf_with_pdfplumber(file_path: str) -> list[Document]:
    docs = []

    with pdfplumber.open(file_path) as pdf:
        for page_index, page in enumerate(pdf.pages):
            text = page.extract_text() or ""

            docs.append(
                Document(
                    page_content=text,
                    metadata={
                        "source": file_path,
                        "page": page_index,
                        "parser": "pdfplumber",
                    },
                )
            )

    return docs


pages = split_documents(load_pdf_with_pdfplumber(PDF_PATH))

print("청크의 수:", len(pages))

pages[10]
```

```python
from langchain_core.output_parsers import JsonOutputParser

# JSON 출력 파서 불러오기
parser = JsonOutputParser()
instructions = parser.get_format_instructions()
print(instructions) # JSON 형식의 지침 출력
```

```python
# 파싱
ai_response = '{"이름":"김철수", "나이":30}'
parsed_response = parser.parse(ai_response)
print(parsed_response)
```

```python
# 프롬프트와 함께 파싱
from langchain_classic.output_parsers import RetryWithErrorOutputParser
from langchain_core.output_parsers import JsonOutputParser
from langchain_openai import ChatOpenAI

# 파서 설정
parser = RetryWithErrorOutputParser.from_llm(
    parser=JsonOutputParser(), 
    llm=ChatOpenAI
    (
        model="unsloth/gemma-4-e2b-it",
        base_url="http://...:.../v1",
        api_key="lm-studio",
    )
)
```

```python
question = "가장 큰 대륙은?"
ai_response = "아시아입니다." # JSON 형식이 아닌 잘못된 응답

try:
    result = parser.parse_with_prompt(ai_response, question)
    print(result)
except Exception as e:
    print(f"오류 발생: {e}")
    # 여기서 AI에게 다시 질문할 수 있습니다.
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

analysis_prompt = ChatPromptTemplate.from_template("이 대답을 영어로 번역해주세요: {answer}")
```

```python
prompt = ChatPromptTemplate.from_template("주제 {topic}에 대해 짧은 설명을 해주세요.")
```

```python
model = ChatOpenAI( 
    model="unsloth/gemma-4-e2b-it",
    base_url="http://...:.../v1",
    api_key="lm-studio",
)
```

```python
parser = StrOutputParser()
```

```python
chain = prompt | model | parser
```

```python
# 이전에 정의된 체인(chain과 새로운 작업을 연결하는 체인 구성
composed_chain = {"answer": chain} | analysis_prompt | model | parser

composed_chain.invoke({"topic":"더블딥"})
```

```python
# (방법 1)여러 작업을 순차적으로 .pipe를 통해 연결하여 체인 구성하기
composed_chain_with_pipe = (
    chain
    .pipe(lambda input: {"answer": input})
    .pipe(analysis_prompt)
    .pipe(model)
    .pipe(parser)
)

composed_chain_with_pipe.invoke({"topic": "더블딥"})
```

```python
# (방법 2) 좀 더 간단하게 연결하기
composed_chain_with_pipe_simple = chain.pipe(lambda input:{"answer": input}, analysis_prompt, model, parser)

composed_chain_with_pipe_simple.invoke({"topic": "더블딥"})
```

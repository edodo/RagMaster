```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel
```

```python
model = ChatOpenAI( 
    model="unsloth/gemma-4-e2b-it",
    base_url="http://...:.../v1",
    api_key="lm-studio",
)
```

```python
parser = StrOutputParser()
```

```python
# 한국어 설명 생성 프롬프트 체인
kor_chain = (
    ChatPromptTemplate.from_template("주제 {topic}에 대해 짧은 설명을 해주세요.")
    | model
    | parser
)

# 영어 설명 생성 프롬프트 체인
eng_chain = (
    ChatPromptTemplate.from_template("주제 {topic}에 대해 짧게 영어로 설명을 해주세요.")
    | model
    | parser
)
```

```python
# 병렬 실행을 위한 RunnableParallel 설정
parallel_chain = RunnableParallel(kor=kor_chain, eng=eng_chain)
```

```python
# 주제에 대한 한국어와 영어 설명 생성
result = parallel_chain.invoke({"topic": "더블딥"})
```

```python
# 결과 출력
print("한글 설명:", result['kor'])
print("영어 설명:", result['eng'])
```

```python
# 자동 대화 이력 관리
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
```

```python
model = ChatOpenAI( 
    model="unsloth/gemma-4-e2b-it",
    # model="qwen3-30b-a3b-instruct-2507",
    base_url="http://...:.../v1",
    api_key="lm-studio",
)
```

```python
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.chat_history import InMemoryChatMessageHistory

chat_history = InMemoryChatMessageHistory()

# 시스템 메시지와 대화 이력을 사용하는 프롬프트 템플릿 정의
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "당신은 금융 상담사입니다. 모든 질문에 최선을 다해 답변하십시오."),
        ("placeholder","{chat_history}"), # 대화 이력 추가
        ("human", "{input}"),
    ]
)

# 대화 이력을 관리할 체인 설정
chain = prompt | model
```

```python
# RunnableWithMessageHistory 클래스를 사용해 체인을 감쌉니다.
chain_with_message_history = RunnableWithMessageHistory(
    chain,
    lambda session_id: chat_history, # 세션 ID에 따라 대화 이력을 불러오는 함수
    input_messages_key="input", # 입력 메시지의 키 설정
    history_messages_key="chat_history", # 대화 이력의 키 설정
)
```

```python
# 질문 메시지 체인 실행
chain_with_message_history.invoke(
    {"input": "저축을 늘리기 위해 무엇을 할 수 있나요?"},
    {"configurable": {"session_id":"unused"}},
).content
```

```python
# 새로운 입력 메시지를 추가하고 체인 실행
chain_with_message_history.invoke(
    {"input": "내가 방금 뭐라고 했나요?"},     # 사용자의 질문
    {"configurable": {"session_id":"unused"}} # 세션 ID 설정
).content
```

```python
# LangChainDeprecationWarning 처리
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

from langgraph.graph import StateGraph, START, END, MessagesState
from langgraph.checkpoint.memory import InMemorySaver
```

```python
# 1. 모델 설정
model = ChatOpenAI(
    model="unsloth/gemma-4-e2b-it",
    # model="qwen3-30b-a3b-instruct-2507",
    base_url="http://...:.../v1",
    api_key="lm-studio",
)
```

```python
# 2. 시스템 메시지와 대화 메시지를 사용하는 프롬프트 정의
# LangGraph의 MessagesState는 대화 전체를 기본적으로 messages라는 state에 누적
# 따라서 프롬프트도 아래처럼 바꾸는 게 자연스러움
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "당신은 금융 상담사입니다. 모든 질문에 최선을 다해 답변하십시오."),
        ("placeholder", "{messages}"),
    ]
)
```

```python
# 3. 기존 chain 유지
chain = prompt | model
```

```python
# 4. LangGraph 노드 함수 정의
def call_model(state: MessagesState):
    response = chain.invoke(
        {
            "messages": state["messages"]
        }
    )

    return {
        "messages": [response]
    }
```

```python
# 5. 그래프 구성
builder = StateGraph(MessagesState)

builder.add_node("call_model", call_model)
builder.add_edge(START, "call_model")
builder.add_edge("call_model", END)
```

```python
# 6. 메모리 저장소 설정
checkpointer = InMemorySaver()

graph = builder.compile(checkpointer=checkpointer)
```

```python
config = {
    "configurable": {
        "thread_id": "unused"
    }
}


# 첫 번째 질문
response = graph.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": "저축을 늘리기 위해 무엇을 할 수 있나요?"
            }
        ]
    },
    config,
)

response["messages"][-1].content
```

```python
# 두 번째 질문
response = graph.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": "내가 방금 뭐라고 했나요?"
            }
        ]
    },
    config,
)

response["messages"][-1].content
```
